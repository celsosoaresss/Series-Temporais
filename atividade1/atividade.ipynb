```python
# Pré-processamento
df_uf = df.copy()
df_uf['Ano'] = pd.to_datetime(df_uf['Ano'], format='%Y')

for uf in df_uf['UF'].unique():
    serie = df_uf[df_uf['UF'] == uf].set_index('Ano')['Quantidade (mil t)'].sort_index()
    
    print(f"\n{'='*60}\nANÁLISE EXPLORATÓRIA E ESTATÍSTICA: {uf}\n{'='*60}")
    
    # 1. Estatística Descritiva
    print("Estatísticas Descritivas:")
    print(serie.describe()[['count', 'mean', 'std', 'min', 'max']])
    
    # 2. Análise de Tendência e Sazonalidade (Decomposição)
    # period=1 reflete que o ciclo é anual
    dec = seasonal_decompose(serie, model='additive', period=2)
    
    fig_dec, axes_dec = plt.subplots(2, 1, figsize=(10, 6), sharex=True)
    axes_dec[0].plot(dec.observed, label='Original', color='blue')
    axes_dec[0].plot(dec.trend, label='Tendência', linestyle='--', color='red')
    axes_dec[0].set_title(f"Decomposição de Tendência - {uf}")
    axes_dec[0].legend()
    
    axes_dec[1].plot(dec.resid, label='Resíduo (Ruído)', color='gray')
    axes_dec[1].set_title("Resíduos / Irregularidades")
    axes_dec[1].legend()
    plt.tight_layout()
    plt.show()

    # 3. Teste de Estacionaridade e Diferenciação
    p_ini = adfuller(serie)[1]
    
    if p_ini > 0.05:
        serie_diff = serie.diff().dropna()
        p_diff = adfuller(serie_diff)[1]
        ordem = 1
        
        if p_diff > 0.05:
            serie_diff = serie_diff.diff().dropna()
            p_diff = adfuller(serie_diff)[1]
            ordem = 2
            
        print(f"Série NÃO estacionária (p={p_ini:.4f}). Estacionarizada com Ordem {ordem} (p={p_diff:.4f}).")
        
        # Comparação Antes e Depois
        fig, ax = plt.subplots(1, 2, figsize=(15, 4))
        ax[0].plot(serie, marker='o', color='blue')
        ax[0].set_title(f"Original (Não Estacionária) - {uf}")
        ax[1].plot(serie_diff, marker='s', color='orange')
        ax[1].set_title(f"Diferenciada (Estacionária d={ordem}) - {uf}")
        plt.show()
    else:
        print(f"Série já é ESTACIONÁRIA (p={p_ini:.4f}).")
```
